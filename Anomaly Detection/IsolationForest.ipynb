```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
import numpy as np
```

```python
df = pd.read_csv("healthcare_outlier_data.csv")
df.head(3)
```

```python
plt.scatter(df.iloc[:,0] , df.iloc[:,1])
```

```python
clf = IsolationForest(contamination="auto")  #it is a threshold 
clf.fit(df)
prediction = clf.predict(df)
```

```python
prediction
```

```python
# -1 matlab outliers
```

```python
index = np.where(prediction < 0)
index
```

```python
x = df.values
```

```python
index = np.where(prediction < 0)
plt.scatter(df.iloc[:,0] , df.iloc[:,1])
plt.scatter(x[index , 0] , x[index , 1] , edgecolors="r")
```
