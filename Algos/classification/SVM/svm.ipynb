```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv('placement_data_scaled.csv')
dataset.head(3)
```

```python
sns.scatterplot(x = 'CGPA' , y= 'TestScore' , data= dataset , hue='Placed')
plt.show()
```

```python
# since it is perfectly linearly separable as we can see in the graph 
# so we should use Simple svm
```

```python
x = dataset.iloc[: , :-1]
y =dataset['Placed']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.20,random_state=15)
```

```python
from sklearn.svm import SVC  #support vector classifier
sv = SVC(kernel='linear')
```

```python
sv.fit(x_train , y_train)
```

```python
sv.score(x_test , y_test)
```

```python
#check overfitting
sv.score(x_train , y_train)
```

```python
# is case me hm random state ko kam ziada kar ke overfitting wagera ko deal kar lege
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy() , clf=sv)
plt.show()
```
