```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset =  pd.read_csv('age_experience_salary.csv')
dataset.head(10)
```

```python
#check null values
dataset.isnull().sum()
```

```python
age = dataset['Age']
salary = dataset['Salary']
exp = dataset['Experience']
age ,salary
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x=age , y=salary)
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x=exp , y=salary)
plt.show()
```

```python
x = dataset.iloc[: , :-1]
y = dataset['Salary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train , y_train)
```

```python
#chk the accuracy
lr.score(x_test , y_test)
```

```python
lr.coef_
#so this is m1 and m2
```

```python
lr.intercept_
#this is c
```

```python
#so now formula become this
# y = 790.82 * age + 2926 * exp + 2989
```

```python
lr.predict(x_test)
```
