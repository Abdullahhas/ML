```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('gpa_vs_package.csv')
dataset.head(3)
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x='GPA' , y ='Package(LPA)' ,data=dataset )
plt.show()
```

```python
x = dataset[['GPA']]
y = dataset['Package(LPA)']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.20,random_state=18)
```

```python
from sklearn.svm import SVR
sv = SVR(kernel='poly')  # we can also use other kernal functions if accuracy is low with the current one
```

```python
sv.fit(x_train , y_train)
```

```python
sv.score(x_test , y_test)
```

```python
sv.score(x_train , y_train)
```

```python

sns.scatterplot(x='GPA' , y ='Package(LPA)' ,data=dataset )
plt.scatter(dataset['GPA'] , sv.predict(x) , color = 'red')
plt.show()
```
