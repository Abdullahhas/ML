```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('gpa_vs_package.csv')
dataset.head(10)
```

```python
dataset.isnull().sum()
```

```python
x = dataset[['GPA']]
x.ndim
```

```python
y = dataset['Package(LPA)']
y.ndim
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='GPA' , y='Package(LPA)' , data= dataset)
plt.show()
```

```python
#split train and test data
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train , y_train)
```

```python
lr.predict([[2.75]])
```

```python
#gives us the accuracy 
lr.score(x_test , y_test)*100
```

```python
y_prd = lr.predict(x)
y_prd
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='GPA' , y='Package(LPA)' , data= dataset)
plt.plot(x , y_prd , c='red')
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='GPA' , y='Package(LPA)' , data= dataset)
plt.plot(dataset['GPA'] , y_prd , c='red')
plt.show()
```
