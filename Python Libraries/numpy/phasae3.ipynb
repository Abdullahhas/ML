```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure : [restaurant_id , 2021 , 2022 , 2023 , 2024]
sales_data = np.array([
        [1, 150000, 180000, 220000, 250000],  # Paradise Biryani,
        [2, 120000, 140000, 160000, 190000],  # Beijing Bites,
        [3, 200000, 230000, 260000, 300000],  # Pizza Hub,
        [4, 180000, 210000, 240000, 270000],  # Burger Point,
        [5, 160000, 185000, 205000, 230000]   # Chai Point,
    ])

print(" ==== Sales Analysis ====")
```

```python
print("Sales Data shape :" , sales_data.shape)
```

```python
print("Sample data for 1st three records : ", sales_data[:3])
```

```python
print("Sample data for 1st three records : ", sales_data[:,1:])  #means row sari le ao pr 1st col se start karo
```

total sales per year

```python
print(np.sum(sales_data , axis = 0))
```

```python
yearly_sum = np.sum(sales_data[: , 1:] , axis=0)
print(yearly_sum)
```

minimum sales per resturanat

```python
min_sales = np.min(sales_data[: , 1:] , axis=1)
print(min_sales)
```

maximum sales per year

```python
max_sales_year = np.max(sales_data[: , 1 :] , axis=0)
print(max_sales_year)
```

Average sales per resutrant

```python
average_sales = np.average(sales_data[: , 1 :] , axis = 1)  # or also used np.mean
print(average_sales)
```

```python
cumsum = np.cumsum(sales_data[: , 1:] , axis=1)
print(cumsum)
```

```python
plt.figure(figsize=(7 ,9))
plt.plot(np.mean(cumsum , axis = 0))
plt.title("Average cumulative sales across all resturants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
