```python
#another way to add
df['Total'] = df.iloc[: , 4:10].sum(axis=1)   #means 10 is excluded and axis =1 means to add horizontally and if axis = 0 meas it add vertically
df.head(3)
```

```python
# write in to csv 

df.to_csv('modified.csv' , index=False)

df.to_csv('modified.csv' , index=False , sep='\t')
```

```python
# get data based on condition
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
```

```python
#filtering textual data
df.loc[df['Name'].str.contains('Mega')]  

df.loc[~df['Name'].str.contains('Mega')]   #  ~ means not equal to
```
